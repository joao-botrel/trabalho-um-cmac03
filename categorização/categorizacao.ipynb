Neste JupyterNotebook, foi feito a discretização baseada em quantis nos dados númericos.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("dataset.csv")
```

```python
df.head(20)
```

```python
#Removendo os dados sem normalização
df = df.drop('Renda per capita', axis=1)
df = df.drop('Valor Total dos bens familiares', axis=1)
df = df.drop('Despesas per capita', axis=1)
```

```python
df.head()
```

```python
#Removendo todos os dados que não vão passar pela discretização

df2 = df.drop('ID-modifed', axis=1)
df2 = df2.drop('Qual a situação da MORADIA DO GRUPO FAMILIAR?', axis=1)
df2 = df2.drop('Qual o principal meio de transporte que você utiliza para vir até a Universidade?', axis=1)
df2 = df2.replace(',', '.', regex=True)
df2 = df2.astype(float)
```

```python
df2.head()
```

```python

#Verificar se há valores NaN em cada coluna
nan_values = df2.isna().sum()

print("Valores NaN em cada coluna:")
print(nan_values)
```

```python
#Verificar a frequência dos valores em cada coluna
value_1 = df2['NORMALIZAÇÃO: RENDA PERCAPITA'].value_counts()

print(f"Frequência dos valores na coluna '{'NORMALIZAÇÃO: RENDA PERCAPITA'}':")
print(value_1)
```

```python
value_2 = df2['NORMALIZAÇÃO: VALOR TOTAL DOS BENS'].value_counts()

print(f"Frequência dos valores na coluna '{'NORMALIZAÇÃO: VALOR TOTAL DOS BENS'}':")
print(value_2)
```

```python
value_3 = df2['Normalização : Despesas per capita'].value_counts()

print(f"Frequência dos valores na coluna '{'Normalização : Despesas per capita'}':")
print(value_3)
```

```python
#Discretização baseada em quantis
def discretize_quantile(series, num_quantiles):
    return pd.qcut(series[series !=0], q=num_quantiles, labels=False, duplicates='drop') + 1
#A discretização só vai ser feita para valores diferentes de zero


num_quantiles = 19

df2['NORMALIZAÇÃO: RENDA PERCAPITA' + ' EM CATEGORIAS'] = discretize_quantile(df2['NORMALIZAÇÃO: RENDA PERCAPITA'], num_quantiles)
df2['NORMALIZAÇÃO: VALOR TOTAL DOS BENS' + ' EM CATEGORIAS'] = discretize_quantile(df2['NORMALIZAÇÃO: VALOR TOTAL DOS BENS'], 14) #Exceção
df2['Normalização : Despesas per capita' + ' EM CATEGORIAS'] = discretize_quantile(df2['Normalização : Despesas per capita'], num_quantiles)

df2.head(50)
```

```python
max1 = df2['NORMALIZAÇÃO: RENDA PERCAPITA EM CATEGORIAS'].max()
max2 = df2['NORMALIZAÇÃO: VALOR TOTAL DOS BENS EM CATEGORIAS'].max()
max3 = df2['Normalização : Despesas per capita EM CATEGORIAS'].max()

print(max1)
print(max2)
print(max3)
```

```python
#Fazendo a categoria para os dados que são igual a 0

df2['NORMALIZAÇÃO: RENDA PERCAPITA EM CATEGORIAS'] = df2['NORMALIZAÇÃO: RENDA PERCAPITA EM CATEGORIAS'].fillna(df2['NORMALIZAÇÃO: RENDA PERCAPITA EM CATEGORIAS'].max() + 1)
df2['NORMALIZAÇÃO: VALOR TOTAL DOS BENS EM CATEGORIAS'] = df2['NORMALIZAÇÃO: VALOR TOTAL DOS BENS EM CATEGORIAS'].fillna(df2['NORMALIZAÇÃO: VALOR TOTAL DOS BENS EM CATEGORIAS'].max() + 1)
df2['Normalização : Despesas per capita EM CATEGORIAS'] = df2['Normalização : Despesas per capita EM CATEGORIAS'].fillna(df2['Normalização : Despesas per capita EM CATEGORIAS'].max() + 1)

df2.head(50)
df2.to_excel('categorias.xlsx', index=False) 
```
