Neste JupyterNotebook foi feito o cálculo do grau de similaridade dos dados númericos e categóricos e após isso a matriz final utilizada para plotar o grafo.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("datasetfinal.csv")
```

```python
df.head(50)
```

```python
df1 = df.drop(columns=['NORMALIZAÇÃO: RENDA PERCAPITA EM CATEGORIAS'])
df1 = df1.drop(columns=['NORMALIZAÇÃO: VALOR TOTAL DOS BENS EM CATEGORIAS'])
df1 = df1.drop(columns=['Normalização : Despesas per capita EM CATEGORIAS'])
df1.head()
```

```python
#Criando a matriz de adjacências para os dados de string
num_alunos = len(df1)
adj_matrix = np.zeros((num_alunos, num_alunos), dtype=float)
elementos = 2

for i in range(num_alunos):
    for j in range(num_alunos):
        if i != j:  # Para não comparar um aluno consigo mesmo
            soma = sum(df1.iloc[i, 1:] == df1.iloc[j, 1:])
            num_atributos_comuns = soma/elementos
            adj_matrix[i][j] = num_atributos_comuns


adj_df = pd.DataFrame(adj_matrix, index=df1.index, columns=df1.index)
adj_df.to_csv('matriz_string.csv', index_label='Aluno')
```

```python
df2 = df.drop(columns=['Qual a situação da MORADIA DO GRUPO FAMILIAR?'])
df2 = df2.drop(columns=['Qual o principal meio de transporte que você utiliza para vir até a Universidade?'])
df2.head()
```

```python
# Criando a matriz de adjacências
num_alunos = len(df2)
adj_matrix = np.zeros((num_alunos, num_alunos), dtype=float)
elementos = 3

for i in range(num_alunos):
    for j in range(num_alunos):
        if i != j: 
            soma = sum(df2.iloc[i, 1:] == df2.iloc[j, 1:])
            num_atributos_comuns = soma/elementos
            adj_matrix[i][j] = num_atributos_comuns

adj_df = pd.DataFrame(adj_matrix, index=df2.index, columns=df2.index)
adj_df.to_csv('matriz_numericos.csv', index_label='Aluno')
```

```python
#Carregando as duas matrizes de adjacências
matriz1 = pd.read_csv('matriz_string.csv', index_col='Aluno')
matriz2 = pd.read_csv('matriz_numericos.csv', index_col='Aluno')

#Criando a matriz de adjacências final
num_alunos = len(df2)
adj_matrix = np.zeros((num_alunos, num_alunos), dtype=float)


for i in range(num_alunos):
    for j in range(num_alunos):
        if i != j: 
            adj_matrix[i][j] = matriz1.iloc[i, j] + matriz2.iloc[i, j]


adj_df = pd.DataFrame(adj_matrix, index=df2.index, columns=df2.index)
adj_df.to_excel('matriz_final.xlsx', index_label='Aluno')
```
